```python
def year(yyyy):
    filtered_df = df[df['學年度'] == yyyy]
    student_counts = filtered_df.groupby(['學制', '學制/系科']).size().reset_index(name='Count')
    return student_counts
```

### make the function

```python
def dataFormater(code_name, data_name, file_name):
    """
    data_name: the name of dataFrame pandas
    file_name: the name based on the year ex, 111.txt
    """
    lis = data_name['學制'].to_list()
    max = len(data_name['學制'].to_list())
    
    f = open(f"./txt_files/{file_name}", "a")
    f.write(f'NAME CODE: {code_name}\n')
    f.write(f'TOTAL ROW: {len(data_name["學制"])}\n')
    f.write(f'TOTAL UNIQUE: {len(data_name["學制"].unique())}\n')
    
    for n in range(len(lis)):
        f.write('{name: "' + f'{data_name["學制/系科"][n]}' + '", value: ' + f'{data_name["Count"][n]}' + '},\n')
    
        if data_name["學制"][n] != data_name["學制"][n+1]:
            f.write('name: "' + f'{data_name["學制"][n]}"\n')
        
        if n == max-2:
            f.write('{name: "' + f'{data_name["學制/系科"][max-1]}' + '", value: ' + f'{data_name["Count"][max-1]}' + '},\n')
            f.write('name: "' + f'{data_name["學制"][max-1]}"\n')
            break
        
    f.write("""
    THE FORMAT IS
    [{
        name: "...",
        children: [
        { name: "...", value: 2 }
        ]
    },
    {
        name: "...",
        children: [
        { name: "...", value: 2 }
        ]
    },] """)   
    f.close()
```
